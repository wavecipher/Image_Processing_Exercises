```python
#import the needed libraries
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

# Exercise 2 : Image Enhancement

The visibility of lane markings, road signs, and obstacles on the roads is significantly reduced at night time. To assist drivers in dark conditions, we can perform contrast enhancement on images captured by the car's front-facing camera and display the enhanced images to the driver. For each image please perform:

---

a) Histogram of the original image's grayscale values.

---


b) Global histogram equalization to the original image.

---


c) Locally adaptive histogram equalization to the original image.

```python
#a
#Load the images
dark_road_1 = cv2.imread('/content/hw1_dark_road_1.jpg')
dark_road_2 = cv2.imread('/content/hw1_dark_road_2.jpg')
dark_road_3 = cv2.imread('/content/hw1_dark_road_3.jpg')

#Convert to grayscale
gray_dark_road_1 = cv2.cvtColor(dark_road_1, cv2.COLOR_BGR2GRAY)
gray_dark_road_2 = cv2.cvtColor(dark_road_2, cv2.COLOR_BGR2GRAY)
gray_dark_road_3 = cv2.cvtColor(dark_road_3, cv2.COLOR_BGR2GRAY)

#I decide to plot the images for better visualization
#Plot the images
plt.figure(figsize=(15, 10))
plt.subplot(2, 3, 1)
plt.imshow(gray_dark_road_1, 'gray')
plt.title('Dark road 1')

plt.subplot(2, 3, 2)
plt.imshow(gray_dark_road_2, 'gray')
plt.title('Dark road 2')

plt.subplot(2, 3, 3)
plt.imshow(gray_dark_road_3, 'gray')
plt.title('Dark road 3')
```

```python
#Histogram of the original images grayscale values

histogram1, bin_edges = np.histogram(gray_dark_road_1, bins=256, range=(0, 255))
histogram2, bin_edges = np.histogram(gray_dark_road_2, bins=256, range=(0, 255))
histogram3, bin_edges = np.histogram(gray_dark_road_3, bins=256, range=(0, 255))

# Create a new figure to display the histogram
plt.figure(figsize=(18, 10))
plt.subplot(2, 3, 1)
plt.title("Grayscale Histogram of 1st image")
plt.xlabel("Grayscale Value")
plt.ylabel("Number of Pixels")
# Set the x-axis limits to ensure the histogram covers the entire grayscale range
plt.xlim([0.0, 255])
# Plot the histogram using the specified bin edges and corresponding values
plt.plot(bin_edges[0:-1], histogram1)


plt.subplot(2, 3, 2)
plt.title("Grayscale Histogram of 2nd image")
plt.xlabel("Grayscale Value")
plt.ylabel("Number of Pixels")
# Set the x-axis limits to ensure the histogram covers the entire grayscale range
plt.xlim([0.0, 255])
# Plot the histogram using the specified bin edges and corresponding values
plt.plot(bin_edges[0:-1], histogram2)


plt.subplot(2, 3, 3)
plt.title("Grayscale Histogram of 3rd image")
plt.xlabel("Grayscale Value")
plt.ylabel("Number of Pixels")
# Set the x-axis limits to ensure the histogram covers the entire grayscale range
plt.xlim([0.0, 255])
# Plot the histogram using the specified bin edges and corresponding values
plt.plot(bin_edges[0:-1], histogram3)
```

```python
#b
#Apply global histogram equalization to the original image
equ_road_1 = cv2.equalizeHist(gray_dark_road_1)
equ_road_2 = cv2.equalizeHist(gray_dark_road_2)
equ_road_3 = cv2.equalizeHist(gray_dark_road_3)

#Plot the images
plt.figure(figsize=(15, 10))
plt.subplot(2, 3, 1)
plt.imshow(equ_road_1, 'gray')
plt.title('Equalized Dark road 1')

plt.subplot(2, 3, 2)
plt.imshow(equ_road_2, 'gray')
plt.title('Equalized Dark road 2')

plt.subplot(2, 3, 3)
plt.imshow(equ_road_3, 'gray')
plt.title('Equalized Dark road 3')
```

```python
#Calculate the histogram of the image after histogram equalization
hist_equ1, bin_edges = np.histogram(equ_road_1, bins=256, range=(0, 255))
hist_equ2, bin_edges = np.histogram(equ_road_2, bins=256, range=(0, 255))
hist_equ3, bin_edges = np.histogram(equ_road_3, bins=256, range=(0, 255))

#Plot the histograms
plt.figure(figsize=(17, 10))
plt.subplot(2, 3, 1)
plt.title("Grayscale Equalized Histogram of 1st image")
plt.xlabel("Grayscale Value")
plt.ylabel("Number of Pixels")
plt.xlim([0.0, 255])
plt.plot(bin_edges[0:-1], hist_equ1)

plt.subplot(2, 3, 2)
plt.title("Grayscale Equalized Histogram of 2nd image")
plt.xlabel("Grayscale Value")
plt.ylabel("Number of Pixels")
plt.xlim([0.0, 255])
plt.plot(bin_edges[0:-1], hist_equ2)

plt.subplot(2, 3, 3)
plt.title("Grayscale Equalized Histogram of 3rd image")
plt.xlabel("Grayscale Value")
plt.ylabel("Number of Pixels")
plt.xlim([0.0, 255])
plt.plot(bin_edges[0:-1], hist_equ3)
```

```python
#c
#Apply locally adaptive histogram equalization to the original image

#Create a Contrast Limited Adaptive Histogram Equalization (CLAHE) object
#Adjust parameters like 'clipLimit' and 'tileGridSize' based on the specific problem
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(10, 10))

#Apply CLAHE algorithm to enhance local contrast
clahe_road_1 = clahe.apply(gray_dark_road_1)
clahe_road_2 = clahe.apply(gray_dark_road_2)
clahe_road_3 = clahe.apply(gray_dark_road_3)

#Calculate the histogram of the image after applying CLAHE
hist_clache1, bin_edges = np.histogram(clahe_road_1, bins=256, range=(0, 255))
hist_clache2, bin_edges = np.histogram(clahe_road_2, bins=256, range=(0, 255))
hist_clache3, bin_edges = np.histogram(clahe_road_3, bins=256, range=(0, 255))
```

```python
#Show the images
cv2_imshow(clahe_road_1)
cv2_imshow(clahe_road_2)
cv2_imshow(clahe_road_3)

#Plot histograms
plt.figure(figsize=(17, 10))
plt.subplot(2, 3, 1)
plt.title("Clahe Histogram of 1st image")
plt.xlabel("Grayscale Value")
plt.ylabel("Number of Pixels")
plt.xlim([0.0, 255])
plt.plot(bin_edges[0:-1], hist_clache1)

plt.subplot(2, 3, 2)
plt.title("Clahe Histogram of 2nd image")
plt.xlabel("Grayscale Value")
plt.ylabel("Number of Pixels")
plt.xlim([0.0, 255])
plt.plot(bin_edges[0:-1], hist_clache2)

plt.subplot(2, 3, 3)
plt.title("Clahe Histogram of 3rd image")
plt.xlabel("Grayscale Value")
plt.ylabel("Number of Pixels")
plt.xlim([0.0, 255])
plt.plot(bin_edges[0:-1], hist_clache3)
```
